## **7. Dataframe 데이터 정렬**

### **1) 데이터 값을 기준으로 정렬**




*   .sort_values() : 데이터 값을 기준으로 행과 열을 정렬하는 메소드입니다. 디폴트 값은 인덱스를 기준으로 오름차순 정렬됩니다. NaN(결측값)이 있다면 그 값은 맨 마지막에 위치합니다. 문자열은 ㄱㄴㄷ(한글), abc(영어) 순으로 정렬합니다.
</br></br>

*   변수명[‘특정 칼럼명’].sort_values() : 특정 칼럼명을 추출한 후 그 값을 정렬하는 방법입니다.
</br></br>

*   .sort_values() 파라미터

    by : 특정 칼럼을 기준으로 정렬
    
    ascending : 정렬의 방향 (True: 오름차순, False: 내림차순)

    axis : 정렬할 대상 축 (=0: 인덱스 기준, =1: 칼럼 기준)

ㅤ

```python
# 행과 열 정렬
print(penguin_df['Sex'].sort_values())  # Sex 칼럼을 추출한 뒤 정렬
```

Sex 칼럼을 추출한 뒤 정렬하였습니다. FEMALE, MALE 순으로 정렬되었고 NaN값은 맨 마지막에 위치한 것을 확인할 수 있습니다.

ㅤ

```python
print(penguin_df.sort_values(by=['Sex'])) # Sex 칼럼을 기준으로 정렬
```

Sex 칼럼을 먼저 정렬한 뒤 그것에 맞춰 다른 칼럼의 데이터 값도 정렬되었습니다.

ㅤ

### **2) 인덱스 값을 기준으로 정렬**


*   .sort_index() : 인덱스 값을 기준으로 행과 열을 정렬합니다. 디폴트 값은 인덱스를 기준으로 오름차순 정렬됩니다.
</br>

*   .sort_index()의 파라미터

    ascending : 정렬의 방향 (True: 오름차순, False: 내림차순)

ㅤ

```python
# 인덱스 값을 기준으로 정렬
print(penguin_df.sort_index()) # 인덱스 기준으로 오름차순으로 정렬
```

```python
print(penguin_df.sort_index(ascending=False)) # 내림차순으로 정렬
```

오름차순으로 정렬된 첫번째 코드의 출력값에 반해 ascending=False을 넣은 두번째 코드의 출력값은 큰값에서 작은값으로 정렬되는 내림차순으로 출력되었습니다.

ㅤ

## **8. Dataframe 데이터 선택**

*   슬라이싱 이용 : df[컬럼명], df[시작인덱스:끝인덱스+1], df[시작인덱스명:끝인덱스명]

*   .loc[행, 열] : 라벨의 이름을 이용하여 데이터 선택 (명칭 기반)

*   .iloc[인덱스 번호] : 위치를 나타내는 인덱스 번호 이용하여 데이터 선택 (위치 기반)
</br></br>


> 추가적으로 ix는 최근에 사라졌으니 참고 부탁드립니다.

ㅤ

```python
## 데이터 선택
# 슬라이싱 이용
print(penguin_df['Culmen Length (mm)'])  # Culmen Length (mm)데이터만 출력
```

Culmen Length (mm) 칼럼만 따로 출력되어 그 칼럼의 데이터만을 확인할 수 있습니다.

ㅤ

```python
print(penguin_df[3:12])  # 3부터 11 인덱스의 데이터만 출력
print(penguin_df[:]) # 전체 데이터 출력
print(penguin_df[3:]) # 3부터 끝까지 데이터 출력
print(penguin_df[:12]) # 처음부터 11 인덱스의 데이터 출력
```

ㅤ

```python
# loc
penguin_df.loc[11, 'Body Mass (g)'] # Body Mass 열에 11 인덱스(행)에 라벨의 이름을 이용 (명칭기반)
```

Body Mass (g) 칼럼에 11 인덱스에 위치된 값인 3300을 출력합니다.

이때 11번째 인덱스가 아니라 0부터 113의 인덱스 중 인덱스명이 11인 데이터 값이 해당됩니다. 명칭기반이기 때문입니다.

ㅤ

```python
# iloc
penguin_df.iloc[2, 3]  # 인덱스 번호 이용 (위치기반)
```

위치기반은 첫번째 순서를 0으로 취급합니다. 그래서 2와 3을 입력한다면 이것은 3번째 인덱스와 4번째 칼럼에 해당하는 값을 출력하게 됩니다.

ㅤ

## **11. Dataframe 데이터 결합**

### **1) concat**

*   concat은 concatenation의 약자로 연결을 의미합니다. 즉, 두 개 이상의 데이터프라임 결합할 때 사용하는 메소드입니다. 

*   pd.concat([데이터프라임], [데이터프라임])

*   axis: 어떻게 결합할 건지 결정하는 파라미터입니다.

  - 0: 열 (디폴트), 1: 행

```python
# concat
df11 = pd.DataFrame([1,2,3,4,5])
df22 = pd.DataFrame([6,7,8,9,10])
print(pd.concat([df11, df22]))   # 열로 합치기

print(pd.concat([df11, df22], axis = 1))  # 행으로 합치기
```

ㅤ

```python
df11 = pd.DataFrame({'A' : ['a', 'a', 'a'], 'B' : ['b', 'b', 'b']}, index = [1,2,3])
df22 = pd.DataFrame({'C' : ['c', 'c', 'c'], 'D' : ['d', 'd', 'd']}, index = [4,5,6])
print(pd.concat([df11, df22]))
```

concat은 칼럼과 인덱스를 기준으로 병합하는데 공통된 칼럼이나 인덱스가 없어 NaN을 포함하여 결합하게 됩니다.

ㅤ

```python
df22 = pd.DataFrame({'C' : ['c', 'c', 'c'], 'D' : ['d', 'd', 'd']}, index = [1,2,3])
print(pd.concat([df11, df22], axis = 1))
```

공통된 인덱스가 존재하여 아까와는 다르게 NaN 값이 포함하지 않고 결합할 수 있습니다.

ㅤ

```python
df11 = pd.DataFrame({'A' : ['a', 'a', 'a'], 'B' : ['b', 'b', 'b']}, index = [1,2,3])
df22 = pd.DataFrame({'A' : ['a', 'a', 'a'], 'B' : ['b', 'b', 'b']}, index = [4,5,6])
print(pd.concat([df11, df22]))
```

반대로 공통된 칼럼이 존재하고 열으로 병합하게 되면 NaN 값을 발생하지 않고 합쳐지게 됩니다.

ㅤ

### **2) append**

*   concat과 달리 행으로 결합이 가능합니다. (열로 결합 불가능)

*   데이터프레임1.append(데이터프레임2)

```python
# append 이용
df11 = pd.DataFrame({'A' : ['a', 'a', 'a'], 'B' : ['b', 'b', 'b']}, index = [1,2,3])
df22 = pd.DataFrame({'C' : ['c', 'c', 'c'], 'D' : ['d', 'd', 'd']}, index = [4,5,6])
print(df11.append(df22))
```
